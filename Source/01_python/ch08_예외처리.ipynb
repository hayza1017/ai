## 2.2 예외를 지정한 처리
```
try:
    예외가 발생할 수도
except 예외타입1:
    해당 예외가 발생할 경우 실행할 문장
except 예외타입2:
    해당 예외가 발생할 경우 실행할 문장
```
**여러 예외처리를 할 경우 상위 예외타입은 아래로 내려야함**


````python
# 예외 처리함
# Exception은 모든 예외의 상위이기 때문에 식의 가장 아래에 입력해야함.
while True :
    try:
        x = int(input('100을 나눌 정수를 입력하시오 > '))
        print('입력한 정수는 {}'.format(x))
        print('100을 입력한 정수로 나누면 {:.2f}'.format(100/x))
        break
    except ZeroDivisionError:
        print('0으로 나눌 수 없습니다. 다시 입력하세요.')
    except valueError:
        print('정수가 아닙니다. 다시 입력하세요.')
    except Exception:
        print('원인 모를 예외가 발생했습니다. 다시 입력하세요.')
print('DONE')
````

````python
# 사용자 정의 예외 : Exception 클래스를 상속받은 클래스
class LengthZeroError(Exception):
    '매개변수 길이가 0일때 발생하는 예외'
    def __init__(self, message):
        super().__init__(message) # 상위(부모)클래스의 생성자에서 예외 메세지를 셋팅
````

````python
raise LengthZeroError("매개변수가 0이예요.")
````

````python
def args_print(*data): # 튜플 매개변수
    if len(data)==0:
        raise LengthZeroError("매개변수 길이가 0이면 예외발생(사용자 정의 예외)")
    for item in data:
        print(item, end=' ')
    print('등을 입력하셨습니다')
````

````python
args_print('a','b','c','d')
````

````python
try:
    args_print()
except Exception as e:
    print(e)
````

- 파이썬에서의 추상크래스는 raise를 이용하거나, @abstractmethod를 이용하여 추상메소드를 구현

````python
class Shape:
    def __init__(self):
        # raise NotImplementedError('추상클래스')
        pass
    def calc_area(self):
        raise NotImplementedError('추상메소드')
        
````

````python
class Shape:
    def __init__(self):
        raise NotImplementedError('객체를 생성할 수 없음')
        pass
    def calc_area(self):
        raise NotImplementedError('추상메소드')
````

````python
# 기능 연쇄기능
class Circle(Shape):
    def __init__(self, r=3):
        self.r = r
    def calc_area(self):
        import numpy as np
        return np.pi * (self.r**2)
````

````python
myCircle = Circle(5)
myCircle.calc_area()
````

````python
# 추상 클래스
from abc import abstractmethod, ABC
class Shape(ABC):
    @abstractmethod
    def __init__(self):
        pass
    @abstractmethod
    def calc_area(self):
        pass
````
