```
• 파일이 없거나 쓰기 금지로 인한 오류

• 리스트 또는 튜플의 인덱스를 벗어난 참조로 인한 오류

```

# 2절. try ~ except로 예외처리
## 2.1 try ~ except
```
try:
    예외가 발생할 가능성이 있는 명령어들
except:
    예외가 발생했을 경우 실행할 명령어들
```

````python
# 예외 처리 하지 않은 수식


# 100을 입력받은 정수값을 나눠 출력
x = int(input('정수를 입력하시오 > ')) # ValueError : invalid literal for int() with base 10 : 'a'
#                                                   invalid litteral(잘못된 문자)를 입력할 경우
print('입력한 정수는 {}'.format(x))
print('100을 입력한 정수로 나누면 {:.2f}'.format(100/x))
# ZeroDivisionError : "division by zero"
#                     0을 입력했을 때 예외타입
print('DONE')
````

````python
# try 절 
try:
    x = int(input('정수를 입력하시오 > '))
    print('입력한 정수는 {}'.format(x))
    print('100을 입력한 정수로 나누면 {:.2f}'.format(100/x))
except:
    print('나누기에 유효한 정수가 아닙니다')
print('DONE')
````
