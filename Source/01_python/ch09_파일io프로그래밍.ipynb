````python
class Member:
    def __init__(self, name, age, email, address):
        self.name = name
        self.age = age
        self.email = email
        self.address = address
    def __str__(self):
        return "{},{},{},{}".format(self.name, self.age, self.email, self.address)
    def __eq__(self, other):
        if isinstance(other, Member):
            return self.__str__() == other.__str__()
# return self.name == other.name and self.age == other.age and self.email == other.email and self.address == other.address
        return False
    
    def as_dict(self): # 객체를 딕셔너리로 return
        return {'name':self.name, 'age':self.age, 'email':self.email, 'address':self.address}
    def as_list(self): # 객체를 list로 return
        return [self.name, self.age, self.email, self.address]
````

````python
member = Member("홍길동", 33, 'hong@h.co,kr', '신림동')
member2 = Member("홍길동", 33, 'hong@h.co,kr', '신림동')
print(id(member), id(member2))
print(member)
print(member == member2)
print(member.__eq__(member2))
print(member.as_dict())
print(member.__dict__)
print(member.as_list())
````

### 2. 객체리스트를 json파일로 저장하기(dump)

````python
# 2. 객체리스트를 json파일로 저장하기(dump)
class Member:
    def __init__(self, name, age, email, address):
        self.name = name
        self.age = age
        self.email = email
        self.address = address
    def as_dict(self):
        '객체를 딕셔너리로 바꿔 return (json.dump 함수 사용에 필요)'
#         return {"name":self.name, "age":self.age, "email":self.email, "address":self.address}
        return self.__dict__
    def __str__(self):
        return "{}님 {}살 {}, {} 주거중".format(self.name,self.age,self.email, self.address) 
    def __eq__(self, other):
        '''
        m1.__eq__(m2)호출시 두객체의 속성변수의 값이 모두 일치하는지 여부
        m1 == m2 구현시 m1.__eq__(m2) 자동 호출됨
        '''
        if isinstance(other, Member):
            return self.__str__() == other.__str__()
        else: 
            return False       
````

````python
member1 = Member("홍길동", 20, "j@hco.kr", "서울")
member2 = Member("홍길동", 20, "j@hco.kr", "서울")
print(member1)
print(member2)
print(member1==1)
print(member1==member2)
````

- ch09_customers.txt 파일 내용은 
```
홍길동, 010-9999-9999, k@hong.com, 30, 5, vvip
유길동, 010-5432-2222, y@hong.com, 20, 2, 가칠해
```

### Customer 클래스 작성
    *Customer("홍길동", "010-9999-9999", "k@hong.com", 30, 5, "vvip")
    *__init__, __str__, as_dict(csv파일), to_txt_style(객체를 txext 스타일 파일로 출력하기 위해)

````python
class Customer:
    def __init__(self, name, phone, email, age, grade, etc):
        self.name = name
        self.phone = phone
        self.email = email
        self.age = age
        self.grade = grade
        self.etc = etc
    def as_dict(self):
        return self.__dict__
    def __str__(self): # ***  이름  010-9999-9999  k@hong.com  30  까칠해
        return "{:5}\t{:3}\t{:15}\t{:20}\t{:3}\t{}".format(self.grade*'*',
                                                          self.name,
                                                          self.phone,
                                                          self.email,
                                                          self.age,
                                                          self.etc)
    def to_txt_style(self):
        return " {}, {}, {}, {}, {}, {}".format(self.name,
                                                self.phone,
                                                self.email,
                                                self.age,
                                                self.grade,
                                                self.etc)
````

````python
c1 = Customer("홍길동", "010-9999-9999", "h@h.com", 30, 4, "기타")
print(c1.as_dict())
print(c1)
print(c1.to_txt_style())
````

````python
def to_customer(row): # "홍길동, 010-9999-9999, h@h.com, 30, 4, vvip \n"  => Customer객체
    data = row.strip().split(', ')
    name = data[0]
    phone = data[1]
    email = data[2]
    age   = int(data[3])
    grade = int(data[4])
    etc = data[5]
    return Customer(name, phone, email, age, grade, etc)
````

````python
c = to_customer("홍길동, 010-9999-9999, k@hong.com, 30, 5, vvip\n")
print(c)
````

````python
def fn1_insert_customer_info():
    '사용자로부처 고객정보를 입력받아 customer형 객체를 return'
    import re
    name = input('이름 :') # 이름 패턴
    name_pattern = r'[가-힣]{2,}'
#     print(re.search(name_pattern, name)) _ 이름 출력이 제대로 되는지 확인
    while not re.search(name_pattern, name):
        print('이름으로 다시 입력하시오(한글 2글자 이상)')
        name = input('이름 :')
    phone = input('전화 :')
    email = input('메일 :')
    while True: # 정규표현식
        try:
            age = int(input('나이 :'))
            if (age<=0) | (age>=130):
                raise Exception("나이 범위 이상")
            break
        except:
            print('올바른 나이를 입력하시오')
    try:
        grade = int(input('등급(1~5) :'))
        if grade<1 :
            grade = 1
        if grade>5 :
            grade = 5
    except:
        grade = 1
        print('유효하지 않는 등급 입력시 1등급으로 초기화')
    etc = input('기타 정보 :')
    return Customer(name, phone, email, age, grade, etc)
````

````python
# 1번 입력 test
customer_list = load_custmers()
customer = fn1_insert_customer_info()
customer_list.append(customer)
for customer in customer_list:
    print(customer)
````

### 2. 전체 출력

````python
def fn2_print_customers(customer_list):
    'customer_list를 출력(pdf 41p. 처럼)'
    print("="*70)
    print("{:^70}".format('고객정보'))
    print("-"*70)
    print("{:>5}\t{:3}\t{:^13}\t{:^20} {:3}   {}".format("GRADE",
                                                      "이름",
                                                      "전화",
                                                      "메일",
                                                      "나이",
                                                      "기타"))
    print("-"*70)
    for customer in customer_list:
        print(customer)
````

````python
# 2번 테스트
fn2_print_customers(customer_list)
````

### 3. 종료

````python
def fn3_save_customer_txt(customer_list):
    'customer_list(객체 list)의 내용을 ch09_customers.txt 백업'
    customer_text_list = []
    for customer in customer_list:
        text = customer.to_txt_style()+'\n'
        customer_text_list.append(text)
    with open('Data/ch09_customers.txt', 'w', encoding='utf-8') as f:
        f.writelines(customer_text_list)
    print("백업 후 종료합니다.")
````

````python
# 3번 테스트
fn3_save_customer_txt(customer_list)
````

###  main() 함수

````python
def main():
    global customer_list
    customer_list = load_custmers() # 파일 내용을 custome_list로
    while True:
        print("1. 입력", "2. 전체출력", "3. 백업후 종료", sep="|", end=" ")
        fn = int(input('메뉴선택 : '))
        if fn==1:
            customer = fn1_insert_customer_info() # 입력받은 내용으로 customer 객체 반환
            customer_list.append(customer)
        elif fn==2:
            fn2_print_customers(customer_list)
        elif fn==3:
            fn3_save_customer_txt(customer_list)
            break
````

````python
if __name__=='__main__':
    main()
````
