# 2절. 생성자(`__init__()`) 와 소멸자(`__del__()`)
```
생성자(함수) : __init__(self, 매개변수, etc.) 객체가 생성될 때 자동 실행
            생성시 필요한 코드 (인스턴스 변수 초기화)를 포함
            없으면 default 생성자가 자동 생성
            
소멸자 : __del__(self) 객체가 소멸될(메모리에서 사라질 = 참조한 변수가 없어질)때 자동 실행(호출하는 친구가 아님)
        인스턴스 객체의 레퍼런스 카운트가 0이 될 때 실행
        소멸시 필요한 코드를 포함

```

````python
class Person:
    def __init__(self, name, gender):
        self.name = name      # 인스턴스 변수
        self.gender = gender
        print('Person 생성자 호출(객체 생성 == 클래스형 변수 생성 == 인스턴스 메모리 생성)')
    def __del__(self):
        print('Person 객체 소멸')
    def print_info(self):
        print(f"{self.name}님은 {self.gender}입니다")
    def __str__(self):
        return f"{self.name}은 {self.gender}이죠"
````

### ※ calss 내의 메소드(함수) 종류

- 인스턴스메소드(일반함수) : 첫번째 매개변수에 self를 꼭 명시. 객체를 통해서만 호출할 수 있는 메소드
- 클래스메소드 : 메소드 위에 @classmethod가 선언된 메소드. 첫번째 매개변수에 cls(clz)를 꼭 명시
- static메소드 : 메소드 위에 @staticmethod가 선언된 메소드. 디폴트 매개변수가 없음

````python
class Person:
    academy = 'MBC' #클래스 변수
    def __init__(self, name='홍길동', gender='남자'): # 생성자
        self.name = name
        self.gender = gender
        
    def print_info(self): #인스턴스메소드 : 변수.print_info()로 호출
        print(f'인스턴스메소드(일반함수):{self.academy}의 {self.name}님은 {self.gender}입니다')
        
    def change_info(self, name, gender='남자'): # 인스턴스메소드 : 변수.change_info("이름","성별")
        self.name = name
        self.gender = gender
        print(f'인스턴스메소드(일반함수)를 통해 이름은 {self.name}으로 성별은 {self.gender}로 변경')
#         print(f'인스턴스메소드(일반함수)를 통해 {self.name}과 {self.gender}변경')
    @classmethod
    def change_class_method(clz, academy): # 클래스 메소드 : Person.change_class_method("KBS")
        # clz == Person
        clz.academy = academy # 클래스 영역의 클래스 변수 수정
        print(f'클래스 메소드를 통해 클래스 변수 academy를 {academy}로 수정')
    @staticmethod
    def print_static_method(): # staticmethod는 클래스변수를 이용하는 경우보다 로직이 우선일 경우
        print(f'static 메소드 : {Preson.academy}로 유지')
````

# 4절. 상속과 재정의
- 상속 : 클래스를 재사용하는 방법
- 상속을 이용하면 부모클래스의 모든 속성을 자식클래스로 물려줄 수 있음
- class 자식클래스명(부모클래스명):

````python
class Shape:
    'Shape는 가로와 세로 멤버변수가 있고, 어쩌구 저쩌구 .....'
    def __init__(self, width, height):
        self.width = width
        self.height = height
        print('Shape 객체 생성')
    def __del__(self):
        print('Shape 객체 소멸')
    def __str__(self):
        return f"가로:{self.width}, 세로:{self.height}"
class Rectangle(Shape):
    def print_area(self):
        print(f"넓이 : {self.width * self.height}") # 상속받음
r = Rectangle(3, 4)
print(r)
print(r.__str__())
r.print_area()
del r
````

- override(함수의 재정의) 기능 : 부모클래스의 함수를 자식클래스에서 다시 정의. 함수이름, 매개변수 개수가 동일해야함.
- overload(함수의 중복정의) 불가 : 동일이름의 함수를 매개변수 개수를 달리하여 정의함.

````python
class Person:
    def __init__(self, name, gender):
        print("Person 객체 생성")
        self.name = name
        self.gender = gender
    def __del__(self):
        print('Preson 객체 소멸')
    def print_info(self):
      # print("{}님 {}".format(self.name, self.gender))
        print(self.__str__())
    def __str__(self):
        return "{}님은 {}".format(self.name, self.gender)
class Student(Person):
    def __init__(self, name, gender, major):
#         self.name = name
#         self.gender = gender
        Person.__init__(self, name, gender)
        self.major = major
    def __del__(self):
        pass
    def __str__(self): # 재정의
        return "{}님의 {}이고, 전공은 {}입니다".format(self.name, self.gender, self.major)
````

## Super() 이용
- super(): "부모클래스의"를 의미 (예약어)

````python
class Person:
    def __init__(self, name, gender):
        print("Person 객체 생성")
        self.name = name
        self.gender = gender
    
    def __del__(self):
        print('Preson 객체 소멸')
    
    def print_info(self):
        print("{}님 {}".format(self.name, self.gender))

    def __str__(self):
        return "{}님은 {}".format(self.name, self.gender)

class Student(Person):
    def __init__(self, name, gender, major):
#         Person.__init__(self, name, gender)
        super().__init__(name, gender)
        self.major = major
    def __del__(self):
        pass
    def print_info(self):
        super().print_info()
        print('전공은 {}입니다.'.format(self.major))
    def __str__(self): # 재정의
        return super().__str__() + "이고, 전공은 {}입니다".format(self.major)
````

````python
s = Student("김철수", "남자", "컴퓨터공학")
print(s)
s.print_info()
````

## static변수(정적변수): 여려 객체들 사이에 데이터를 공유하고 싶을 때 사용

* static변수 앞에는 under score 2개(__)를 붙이면, 클래스명._클래스명__static변수명 이름으로 참조
* 클래스변수는 클래스나 인스턴스나 외부에서 바로 접근 가능. 반면 __로 시작하는 변수는 외부에서 클래스명._클래스명__static변명과 같이 접근 해야하므로 외부에서 실수로 접근하거나 변경하는 것을 방지
* static 변수는 크래스 내부에서만 접근하도록 하려는 의도


````python
class Person:
    def __init__(self, name, gender):
        print("Person 객체 생성")
        self.name = name
        self.gender = gender
    
    def __del__(self):
        print('Preson 객체 소멸')
    
    def print_info(self):
        print("{}님 {}".format(self.name, self.gender))

    def __str__(self):
        return "{}님은 {}".format(self.name, self.gender)

class Student(Person):
    __count = 0 # static변수(정적변수, 공유변수) Student형 변수 몇개 생성되었는지 저장
    def __init__(self, name, gender, major):
        super().__init__(name, gender)
        self.major = major
        Student._Student__count += 1 # 객체가 생성될 때마다 __count 1증가
    def __del__(self):
        Student._Student__count -= 1 # 객체가 소멸될 때마다 __count 1감소
    def print_info(self):
        super().print_info()
        print('전공은 {}입니다-이상{}개 생성'.format(self.major, Student._Student__count))
    def __str__(self): # 재정의
        return super().__str__() + "이고, 전공은 {}입니다-이상{}개 생성".format(self.major, Student._Student__count)
    
    @classmethod
    def get_count(cls):
        # return Student._Student__count
        return cls.__count
````

````python
s1 = Student("홍", "남자", "AI")
print(s1)
print(s1.get_count())
print(Student.get_count())
````

````python
# 외부에서 static변수 실수로 수정 못하도록 만드는 용도
Student.__count = 0
Student.get_count()
````

```
        • 삼각형 객체가 몇 개가 만즐어져 있는지 정보를 저장하는 클래스 변수를 추가하고 생성자에서 
          객체를 만들때마다 증가시키고 소멸자에게 객체를 소멸시킬 때마다 감소합니다.(staitc변수 이용)
    ✓ Triangle 객체 생성예제
      t1 = Triangle(4,3,0,10); t2 = Triangle(3, 4, 5, 5)
      print(t1) ; print(t2)
      print('Triangle 객체 객수 :', Triangle.get_count())
      t1.move(50, 50)
      print(t1)
```

````python
class Shape:
    def __init__(self, x=0, y=0):
        print("Shape 객체 생성")
        self.x = x
        self.y = y
    
    def move(self, x, y):
        self.x = x
        self.y = y
        
    def __str__(self):
        return "{}, {} 좌표값".format(self.x, self.y)
    
    @staticmethod
    def static_method():
        print("암거나")

class Triangle(Shape):
    __count = 0 # static변수(정적변수, 공유변수) Student형 변수 몇개 생성되었는지 저장
    def __init__(self, width, height, x=0, y=0):
        super().__init__(x, y)
        self.width = width
        self.height = height
        Triangle._Triangle__count += 1 # 객체가 생성될 때마다 __count 1증가
    def __del__(self):
        Triangle._Triangle__count -= 1 # 객체가 소멸될 때마다 __count 1감소
    def calc_area(self):
        print(f"넓이 : {self.x*self.y*0.5}")
    def __str__(self): # 재정의
        return super().__str__() + "이고, 폭은 {}이고 높이는 {}입니다.".format(self.width, self.height)
    
    @classmethod
    def get_count(cls):
        # return Student._Student__count
        return cls.__count
````

````python
t1 = Triangle(4,3,0,10)
t2 = Triangle(3, 4, 5, 5)
print(t1)
print(t2)
print('Triangle 객체 객수 :', Triangle.get_count())
t1.move(50, 50)
print(t1)
````
