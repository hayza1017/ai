# 2절. 생성자(`__init__()`) 와 소멸자(`__del__()`)
```
생성자(함수) : __init__(self, 매개변수, etc.) 객체가 생성될 때 자동 실행
            생성시 필요한 코드 (인스턴스 변수 초기화)를 포함
            없으면 default 생성자가 자동 생성
            
소멸자 : __del__(self) 객체가 소멸될(메모리에서 사라질 = 참조한 변수가 없어질)때 자동 실행(호출하는 친구가 아님)
        인스턴스 객체의 레퍼런스 카운트가 0이 될 때 실행
        소멸시 필요한 코드를 포함

```

````python
class Person:
    def __init__(self, name, gender):
        self.name = name      # 인스턴스 변수
        self.gender = gender
        print('Person 생성자 호출(객체 생성 == 클래스형 변수 생성 == 인스턴스 메모리 생성)')
    def __del__(self):
        print('Person 객체 소멸')
    def print_info(self):
        print(f"{self.name}님은 {self.gender}입니다")
    def __str__(self):
        return f"{self.name}은 {self.gender}이죠"
````

````python
p1 = Person('심사임당', '여자')
p1.print_info()
print(p1)
print(p1.__str__())
````

````python
type(p1), isinstance(p1, Person)
````

````python
del p1
````

````python
p2 = Person("강호동", "남자")
p3 = p2
````

````python
del p2 # 레퍼런스 카운트가 0이 아님
````

````python
del p3 # 레퍼런스 카운트가 0이 됨
````

## 2. 올바른 생성자 함수식

````python
# 올바른 생성자 함수식
class Person:
    def __init__(self,name="아무개", gender="남자"):
        self.name = name
        self.gender = gender
    def __str__(self):
        return f"{self.name}은 {self.gender}이네요"
````

````python
p1 = Person()
p2 = Person("홍길동")
p3 = Person("신길숙", "여자")
print(p1) # print('p1=', p1)
print(p2) # print('p2=', p2)
print(p3) # print('p3=', p3)
````

## 1. 클래스 변수 : 클래스 내부 선언
- 생성된 모든 인스턴스(객체)에 동일한 값을 가짐
- 클래스.클래스 변수를 수정하면 모든 인스턴스에 반영 될 수 있음
- 객체 변수.클래스 변수를 수정하면 해당 객체에만 반영

````python
class Person:
    academy = 'MBC' # 클래스 변수
    def __init__(self, name="신길동", gender="남자"): # self : 내 객체 자신
        self.name = name
        self.gender = gender
        print('Person 객체 생성합니다')
    def __str__(self): 
        return"%s의 %s은/는 %s입니다" % (self.academy, self.name, self.gender)
print(Person)
print(Person.academy)
````

````python
p1 = Person()
p2 = Person()
````

````python
print(p1)
print(p2)
````

````python
Person.academy = "KBS"
print(p1)
print(p2)
p2.academy = "딴데"
p2.name = "홍길동"
print(p1)
print(p2)
````
