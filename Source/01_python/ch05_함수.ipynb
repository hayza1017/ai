## 3.4 튜플 매개변수를 이용한 가변인수 설정

````python
# 함수 정의
def add(a, b):
    return a+b
def add(a, b, c):
    return a+b+c
````

````python
# args 앞의 * : 매개변수들(인수들)이 args 튜플로 전달
def add(*args):
    "전달받은 매개변수들의 합을 return. 매개변수가 없을 경우 0을 return"
    # print(args) # (1, 2, 3)
    tot = 0 #누적변수
    for num in args:
        tot += num # tot = tot + num
    return tot
````

````python
print(add())
print(add(1))
print(add(10,20))
print(add(10, 2, 3))
print(add(10, 2, 3, 4))
````

## 3.7 인수의 언패킹
### 튜플인수 언패킹
- 패킹 : 함수 정의시 매개변수 앞에 *붙임(튜플매개변수)
- 언패킹 : 함수 호출시 매개변수 앞에 *붙임(리스트나 튜플이 각자 요소별 매개변수로 들어감)

````python
def add(*args):
    tot = 0 #누적변수
    for num in args:
        tot += num # tot = tot + num
    return tot
````

````python
numbers = (1, 2, 3, 4, 5, 6)
add(*numbers) # *앞에 붙이면 나열가능한 것이다
````

````python
def map_template(func, L=[]):
    result = []
    for item in L:
        result.append(func(item))
    return result
````

````python
def filter_template(func, L):
    result = []
    for item in L:
        if func(item):
            result.append(item)
            
    return result
````
