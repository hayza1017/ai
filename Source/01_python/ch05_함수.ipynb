````python
# 렉시컬 특징 : 변수가 어디에 선언되었는지에 따라 변수 범위가 정해짐
g_var = 100 # 전역변수
def func3():
    print('함수안 g_var: ', g_var)
    g_var = 200 # 지역변수
func3()
print('함수 밖 g_var :', g_var)
````

````python
# 렉시컬 특징
g_var = 100 # 전역변수
def func3():
    global g_var # func3함수 내의 g_var는 전역변수를 쓴다
    print('함수안 g_var: ', g_var) # 전역번수
    g_var = 200 # 전역변수
func3()
print('함수 밖 g_var :', g_var)
````

# 2절. 함수의 실행결과 반환하는 return

````python
def my_add_print(a,b):
    "매개변수로 들어온 a,b 값의 합을 출력합니다 - docstring"
    total = a + b
    print("합은 {}입니다".format(total))
def my_add_return(a, b):
    "매개변수로 들어온 a, b값의 합을 return합니다"
    total = a + b
    return "합은 {}입니다".format(total)
````

````python
result = my_add_print(10, 20)
````

````python
print(result)
print(type(result))
````

````python
result = my_add_return(5,7)
print(result)
print(type(result))
````

### 여러개 값을 return

````python
def swap(a, b):
    "매개변수의 순서를 바꿔 return"
    return b, a # 여러개 값을 return하면 tuple로 전달
````

````python
result = swap(10, 20)
print(result)
print(type(result))
````

````python
x, y = 10, 20
print(f"함수 실행전 : x = {x}, y={y}")
x, y = swap(x, y)
print(f"함수 실행후 : x = {x}, y={y}")
````

## 3.3 순서인수, 키워드 인수
- 순서인수 : 함수의 정의된 순서대로 전달 방식
- 키워드인수 : 인수의 이름을 명시하는 방식

````python
def func(a, L=None):
    if L is None:
        L = []
    L.append(a)
    return L
````

````python
lst = []
print(func(10, lst)) # 순서인수
print(func(L=[], a=10)) # 키워드인수
print(func(20)) # 순서인수 a에 20이 전달
print(func(20, None))
````
