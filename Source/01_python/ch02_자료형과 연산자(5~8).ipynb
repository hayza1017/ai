## 5.3 논리연산자: and, &, or, |, not

````python
# & : 논리연산자, 비트연산자
print((10>3) & (10>5)) #논리연산 True & True => True
print( 3 & 10 ) # 비트연산 수 & 수 (비트란_이진수 하나하나)
# 3: 0 0 1 1
# 10: 1 0 1 0
# &: 0 0 1 0 (=2)
print( 10>3 & 10>5) # 10>2>5
````

````python
# and : 논리연산자, 거짓으로 판별되는 첫번째 항의 결과 반환
#       모든 항이 참이면 마지막항을 반환한다.
print( True and True and 'abc')
print(0 and 5)
print(10>3 and 10>5)
````

````python
# | : 논리연산자, 비트연산자
print((10>3) | (10>5)) #논리연산 True & True => True
print( 3 | 10 ) # 비트연산 수 & 수 (비트란_이진수 하나하나)
# 3: 0 0 1 1
# 10: 1 0 1 0
# |: 1 0 1 1 (11)
print( 10>3 | 10>5) # 10>11>5
````

````python
# or : 논리연산자, 참으로 판별되는 첫번째 항의 결과 반환
#       모든 항이 거짓이면 마지막항을 반환한다.
print( False or False or 'abc')
print(0 or '홍길동')
print(10>3 or 10>5)
````

# 6절 문자열 다루기
- 문자열 슬라이싱 변수[from:stop:sstep]
- len(문자변수) : 문자수
- upper() : 대문자로
- lower() : 소문자로
- title() : 각 어절의 앞글자만 대문자로
- capitalize() : 첫문자만 대문자로
- count('찾을 문자열') '찾을 문자열'이 몇개 들어있는지 갯수. 없으면 0
- find('찾을 문자열') : '찾을 문자열'이 왼쪽부터 시작하여 몇번째에 있는지 index. 없으면 -1
- rfind('찾을 문자열') : '찾을 문자열'이 오른쪽부터 시작하여 몇번째에 있는지 index. 없으면 -1
