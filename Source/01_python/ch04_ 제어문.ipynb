## 2.1 for-in문

```
for 임시변수 in 나열가능한 자료:
    반복할 문장들
else:
    나열가능한 자료를 다 소요하여 for블록을 다 수행한 후, 최종 한번 실행할 문장들

```

````python
for i in range(1, 10, 2): #[1, 3, 5, 7, 9]
    print(i, end="\t")
else:
    print("\nDONE")
    print('END')
````

## 2.2 while문
```
while 조건:
    조건이 참일 경우 반복할 문장들
else:
    조건이 거짓일 경우 한번 실행할 문장(else 선택사항)
    

```

````python
# 1~100까지 짝수들의 합을 계산하고 출력하시오 (while문)
tot = 0 # 누적변수
i = 1
while i<=100:
    if i%2==0:
        tot += i
    i += 1 # i 변수 1 증가
#else:
print('1~100까지 짝수의 합은', tot)
````

````python
# ex1_사용자에게 두 정수를 입력받아 작은수부터 큰수까지 홀수들의 합을 누적하여 출력하시오
first = int(input('첫번째 정수는?'))
second = int(input('두번째 정수는?'))
# (1) 두수 중 작은 수는 minNo, 큰수는 maxNo에 할당
if first<second:
    minNo = first
    maxNo = second
else:
    minNo = second
    maxNO = first
# minNO = first if first<second else second
# maxNo = first if first>second else second

# (2) minNo~maxNo까지 홀수 누적하기
tot = 0
for no in range(minNo, maxNo+1):
    if no%2: # if no%2!=0
        tot += no
print(f"{minNo}부터 {maxNo}까지 홀수의 합은 {tot}입니다")
````

````python
# ex1.2_사용자에게 두 정수를 입력받아 작은수부터 큰수까지 홀수들의 합을 누적하여 출력하시오
first = int(input('첫번째 정수는?'))
second = int(input('두번째 정수는?'))
# (1) first에 작은 수를 second에 큰 수
if first>second:
    # 바꿔
    # temp = first
    # first = second
    # second = temp
    first, second = second, first
# (2) first부터 second까지 홀수합 누적
tot = 0
for no in range(first, second):
    if no%2: # if no%2!=0
        tot += no
print(f"{first}부터 {second}까지 홀수의 합은 {tot}입니다")
````

````python
# ex2_1~45까지의 로또번호 6개를 추출하여 정렬된 로또번호를 출력하는 프로그램을 구현하시오
import random
# random.randint(1, 45)를 최소 6이상 => 리스트, 셋-> 정렬된 값 출력
lotto_lst = []
cnt = 0 # 난수 발생 횟수
while len(lotto_lst)<6:
    cnt += 1
    temp = random.randint(1, 45) # 난수
    if temp not in lotto_lst: # temp가 lotto_lst의 요소가 아닌지 여부
        lotto_lst.append(temp)
print('로또번호 :', lotto_lst)
print('정렬된 로또번호 :', sorted(lotto_lst))
print('while문 수행 횟수(난수 발생 횟수) :', cnt)
````

````python
# ex. 1~100사이의 정수를 맞추는 프로그램을 구현하시오
'''
(프로그램 안에서 random수가 만들어지고 그 random수 맞추기)
1단계 : 수를 못 맞추면 UP/DOWN을 출력하고, 맞추면 축하메세지 뿌리고 끝남
(1~100사이의 수를 알아맞추기)
2단계 : 수를 못 맞추면 min~max를 조정, 맞추면 축하메세지 뿌리고 끝남
3단계 : 2단계의 기회를 5회로 제한. 5회의 기회가 지나면 실패메세지 뿌리고 끝남
(2단계와 3단계는 연습문제로)
'''
import random
randomNum = random.randint(1, 100)
print('컴퓨터가 발생한 임의의 수 :', randomNum)
while True :
    inputNum = int(input("1~100 사이의 수를 맞춰 보시오 :")) # 2단계에서
    if (inputNum>100) | (inputNum<1):
        print('수의 범위를 벗어났습니다')
    elif inputNum>randomNum :
        print("DOWN ↓")
    elif inputNum<randomNum :
        print("UP ↑")
    else :
        print('축하합니다')
        break
    print('DONE')
````

````python
# 중첩 루프 탈출 방법1 : flag변수 이용
flag = False
for a in range(0,3): # a <= 0, 1, 2
    for b in range(1,3): # b <= 1, 2
        if a == b:
            print("a={}, b={} break".format(a,b))
            flag = True
            break
        print(a, b)
    if flag:
        break
        
print("DONE")
````

````python
# ex. 1~100사이의 정수를 맞추는 프로그램을 구현하시오
'''
(프로그램 안에서 random수가 만들어지고 그 random수 맞추기)
1단계 : 수를 못 맞추면 UP/DOWN을 출력하고, 맞추면 축하메세지 뿌리고 끝남
(1~100사이의 수를 알아맞추기)
2단계 : 수를 못 맞추면 min~max를 조정, 맞추면 축하메세지 뿌리고 끝남
3단계 : 2단계의 기회를 5회로 제한. 5회의 기회가 지나면 실패메세지 뿌리고 끝남
(2단계와 3단계는 연습문제로)
'''
import random
randomNum = random.randint(1, 100)
print('컴퓨터가 발생한 임의의 수 :', randomNum)

inputNum = 0
min_num = 1
max_num = 100

while inputNum! = randomNum:
    inputNum = int(input(f"{min_num1}~{max_num}사이의 수를 맞춰 보시오 :")) # 2단계에서 min~max사이의 수를 맞춰보심
    if (inputNum>max_num) | (inputNum<min_num):
        print('수의 범위를 벗어났습니다')
    elif inputNum>randomNum:
        print("DOWN ↓")
    # 2단계는 max값 조정
        max_num = inputNum - 1
    elif inputNum<randomNum:
        print("UP ↑")
       #2단계는 min값 조정
        min_num = inputNum + 1
    else:
        print('축하합니다')

    print('DONE')
````

````python
# ex3번. 1~100사이의 정수를 맞추는 프로그램을 구현하시오
'''
(프로그램 안에서 random수가 만들어지고 그 random수 맞추기)
1단계 : 수를 못 맞추면 UP/DOWN을 출력하고, 맞추면 축하메세지 뿌리고 끝남
(1~100사이의 수를 알아맞추기)
2단계 : 수를 못 맞추면 min~max를 조정, 맞추면 축하메세지 뿌리고 끝남
3단계 : 2단계의 기회를 5회로 제한. 5회의 기회가 지나면 실패메세지 뿌리고 끝남
(2단계와 3단계는 연습문제로)
'''
import random
randomNum = random.randint(1, 100)
print('컴퓨터가 발생한 임의의 수 :', randomNum)

inputNum = 0
min_num = 1
max_num = 100

for cnt in range(5, 0, -1):
    inputNum = int(input(f"{min_num1}~{max_num}사이의 수를 맞춰 보시오(남은 기뢰: {cnt}번) :"))
    if (inputNum>max_num) | (inputNum<min_num):
        print('수의 범위를 벗어났습니다')
    elif inputNum>randomNum:
        print("DOWN ↓")
    # 2단계는 max값 조정
        max_num = inputNum - 1
    elif inputNum<randomNum:
        print("UP ↑")
       #2단계는 min값 조정
        min_num = inputNum + 1
    else:
        print('축하합니다')
        break
else:
    print("기회를 모두 사용하셨습니다.")
print('DONE')
````
