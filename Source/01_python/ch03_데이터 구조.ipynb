- index(찾을 데이터, start) : start번째부터 해당 찾을 데이터가 몇번째 index에 있는지 반환

```python
def safe_index(lst, item):
    '''
    첫번째 매개변수 lst에서 itm 요소가 있는 index를 변환하는 함수
    item 요소가 없으면 -1를 반환
    '''
    if item in lst:
        return lst.index(item)
    else:
        return -1
```

```python
def safe_remove(lst, remove_item):
    '''
    첫번째 매개변수 lst에서 remove_ite 데이터 제거
    해당 remove_item이 lst에 없으면 메세지 출력
    '''
    if remove_item in lst:
        lst.remove(remove_item)
    else:
        print('해당 데이터는 없는 데이터입니다.')
```

## 1.9 리스트 복제하기

```python
# 스칼라 변수의 복제
i = 10
copied_i = i
print(f"원본i={i}, 복사본i={copied_i} / {id(i)}, 복사본주소 : {id(copied_i)}")
copied_i += 5 # copied_i = copied_i + 5
print(f"원본i={i}, 복사본i={copied_i} / {id(i)}, 복사본주소 : {id(copied_i)}")
```

```python
# 데이터 구조의 복제(X)
numbers = [5, 1, 9] # 원본
new_numbers = numbers # 복제본 X, 할당한 것임
print(f"원본i={numbers}, 복사본i={new_numbers} / {id(numbers)}, 복사본주소 : {id(new_numbers)}")
new_numbers[0] = 555
print(f"원본i={numbers}, 복사본i={new_numbers} / {id(numbers)}, 복사본주소 : {id(new_numbers)}")
```

```python
# 복제(=깊은 복사)
numbers = [5, 1, 9] # 원본
new_numbers = numbers.copy() # 복제본(깊은 복사)
print(f"원본i={numbers}, 복사본i={new_numbers} / {id(numbers)}, 복사본주소 : {id(new_numbers)}")
new_numbers.sort()
print(f"원본i={numbers}, 복사본i={new_numbers} / {id(numbers)}, 복사본주소 : {id(new_numbers)}")
```

```python
# 복제(=깊은 복사)
numbers = ['김', '이', '박'] # 원본
new_numbers = numbers.copy() # 복제본(깊은 복사)
print(f"원본i={numbers}, 복사본i={new_numbers} / {id(numbers)}, 복사본주소 : {id(new_numbers)}")
new_numbers.sort()
print(f"원본i={numbers}, 복사본i={new_numbers} / {id(numbers)}, 복사본주소 : {id(new_numbers)}")
```
