```python
def safe_index(lst, item):
    '''
    첫번째 매개변수 lst에서 itm 요소가 있는 index를 변환하는 함수
    item 요소가 없으면 -1를 반환
    '''
    if item in lst:
        return lst.index(item)
    else:
        return -1
```

```python
# 데이터 구조의 복제(X)
numbers = [5, 1, 9] # 원본
new_numbers = numbers # 복제본 X, 할당한 것임
print(f"원본i={numbers}, 복사본i={new_numbers} / {id(numbers)}, 복사본주소 : {id(new_numbers)}")
new_numbers[0] = 555
print(f"원본i={numbers}, 복사본i={new_numbers} / {id(numbers)}, 복사본주소 : {id(new_numbers)}")
```

```python
# 복제(=깊은 복사)
numbers = [5, 1, 9] # 원본
new_numbers = numbers.copy() # 복제본(깊은 복사)
print(f"원본i={numbers}, 복사본i={new_numbers} / {id(numbers)}, 복사본주소 : {id(new_numbers)}")
new_numbers.sort()
print(f"원본i={numbers}, 복사본i={new_numbers} / {id(numbers)}, 복사본주소 : {id(new_numbers)}")
```
